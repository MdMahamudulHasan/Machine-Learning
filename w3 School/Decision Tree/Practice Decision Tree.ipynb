```python
# create and display decision tree

import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt

df = pd.read_csv("G:/Machine Learning/w3 School/Decision Tree/data.csv")

d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)

d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']

dtree = tree.DecisionTreeClassifier().fit(X, y)

#plotting the Decision Tree
plt.figure(figsize=(12, 8))
tree.plot_tree(dtree, feature_names= features, filled=True, rounded=True)
plt.title('Decision Tree Visualization')
plt.show()
```

```python
#use predict() method to predict new values

import pandas as pd
from sklearn import tree

df = pd.read_csv("G:/Machine Learning/w3 School/Decision Tree/data.csv")

d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)

d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']

dtree = tree.DecisionTreeClassifier().fit(X, y)

data = pd.DataFrame([[40, 10, 6, 1]], columns = features)
print(dtree.predict(data))
print("[0]-> means 'Go' \n[1]-> means 'No'")

print("\n\n")

data = pd.DataFrame([[40, 10, 7, 1]], columns = features)
print(dtree.predict(data))
print("[0]-> means 'Go' \n[1]-> means 'No'")
```
