```python
# plot Weight vs CO2 with prediction value

import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv("G:/Machine Learning/w3 School/Multiple Regression/data.csv")

X = df[['Volume', 'Weight']]
y = df['CO2']

regr = linear_model.LinearRegression().fit(X, y)

prediction_data = pd.DataFrame({'Volume':[1300],'Weight':[3300]})
predictCO2 = regr.predict(prediction_data)

plt.scatter(df['Weight'], y, color='blue', label='Actual Data: Weight vs CO2')

plt.scatter(prediction_data['Weight'], predictCO2, color='red', label='Predicted CO2')

plt.xlabel('Weight(g)')
plt.ylabel('CO2 Emissions')
plt.title('Weight vs CO2 Emissions with prediction value')

plt.legend()
plt.show()
```

```python
# plot Volume vs CO2 with prediction value
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("G:/Machine Learning/w3 School/Multiple Regression/data.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

newdata= pd.DataFrame({'Weight':[2300], 'Volume': [1300]})
predictionCO2 = regr.predict(newdata)


# plot Volume vs CO2
plt.scatter(df['Volume'], y, color='green', label='Actual Data: Volume vs CO2')

# plot the prediction point
plt.scatter(newdata['Volume'], predictionCO2, color='red', label='Predicted CO2', marker='X', s=100)

#add labels and title
plt.xlabel('Volume(ccm)')
plt.ylabel('CO2 Emissions')
plt.title('Volume vs CO2 Emissions with prediction value')

plt.legend()
plt.show()
```

```python
# plotting Weight and Volume vs CO2

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("G:/Machine Learning/w3 School/Multiple Regression/data.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

newdata= pd.DataFrame({'Weight':[2300], 'Volume': [1300]})
predictionCO2 = regr.predict(newdata)

# Plot Weight vs CO2
plt.scatter(df['Weight'], y, color='blue', label='Actual Data: Weight vs CO2')

# Plot Volume vs CO2 using the same y-axis
plt.scatter(df['Volume'], y, color='green', label='Actual Data: Volume vs CO2')

# Plot the predicted point
plt.scatter(newdata['Weight'], predictionCO2, color='red', label='Predicted CO2', marker='X', s=100)

# Add labels and title
plt.xlabel('Weight (g) and Volume (ccm)')
plt.ylabel('CO2 Emissions')
plt.title('Weight and Volume vs CO2 Emissions with Prediction')

plt.legend()
plt.show()
```

```python

# Volume vs CO2 with regression line
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("G:/Machine Learning/w3 School/Multiple Regression/data.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

new_data= pd.DataFrame({'Weight':[2300], 'Volume': [1300]})
predictedCO2 = regr.predict(new_data)

volume_range = np.linspace(df['Volume'].min(), df['Volume'].max(), 100)

predicted_CO2_volume = regr.predict(pd.DataFrame({'Weight': np.full_like(volume_range, 2300), 'Volume': volume_range}))

plt.scatter(df['Volume'], y, color='green', label='Actual Data: Volume vs CO2', alpha=0.6)
plt.plot(volume_range, predicted_CO2_volume, color='orange', label='Regression Line (Weight=2300)', linewidth=2)

# Plot the predicted point
plt.scatter(new_data['Volume'], predictedCO2, color='red', label='Predicted CO2 (Weight=2300, Volume=1300)', marker='X', s=100)

# Add labels and title
plt.xlabel('Volume (ccm)')
plt.ylabel('CO2 Emissions')
plt.title('Volume vs CO2 Emissions with Regression Line')

# Show legend
plt.legend()

# Display the plot
plt.show()
```
