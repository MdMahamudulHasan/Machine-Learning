```python
# start by drawing a scatter plot

import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

plt.scatter(x, y)
plt.show()
```

```python
# draw polynomialregression line

import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = np.poly1d(np.polyfit(x, y, 3))

myline = np.linspace(min(x), max(x), 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
# gernerate graph for degree of 1, 2, 3, 4

import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

model_1 = np.poly1d(np.polyfit(x, y, 1))
model_2 = np.poly1d(np.polyfit(x, y, 2))
model_3 = np.poly1d(np.polyfit(x, y, 3))
model_4 = np.poly1d(np.polyfit(x, y, 4))

line = np.linspace(min(x), max(x), 100)

plt.scatter(x, y)
plt.plot(line, model_1(line), color='red', label='Degree 1')
plt.plot(line, model_2(line), color='green', label='Degree 2')
plt.plot(line, model_3(line), color='orange', label='Degree 3')
plt.plot(line, model_4(line), color='purple', label='Degree 4')

plt.title("Polynomial Fits of Different Degrees")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.show()
```

```python
# how well does my data fit in a polynomial regression

import numpy as np
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = np.poly1d(np.polyfit(x, y, 3))

print(r2_score(y, mymodel(x)))
```

```python
# predict the speed of a car passing at 17:00

import numpy as np
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = np.poly1d(np.polyfit(x, y, 3))

speed= mymodel(17)

print(speed)
```

```python
# plotting for prediction value

import numpy as np
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel= np.poly1d(np.polyfit(x, y, 3))

myline = np.linspace(min(x), max(x), 100)

#plotting 
plt.scatter(x,y, marker='X',s=20, label='Data points',color='blue')
plt.plot(myline, mymodel(myline), color='purple', label='Polinomial Fist')

# plotting for prediction value of speed
plt.hlines(speed, xmin=min(x), xmax=17, colors='yellow')
plt.vlines(17, ymin=min(y), ymax=speed, colors='yellow')

#adding label
plt.title('Polynomial Fit and prediction for x = 17')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.show()
```
