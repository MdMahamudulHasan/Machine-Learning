```python
# start by visualizing some data points

import matplotlib.pyplot as plt

x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

plt.scatter(x, y)
plt.show()
```

```python
# Now utilize the elbow method to visualize the inertia for different values of K

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

data = list(zip(x, y))

inertias = []

for i in range (1, len(data)+1):
  kmeans = KMeans(n_clusters= i)
  kmeans.fit(data)
  inertias.append(kmeans.inertia_)

plt.plot(range(1, len(data)+1), inertias, marker = 'o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# the elbow method show that 2 is a good value for K, so we retrain and visualize the result:

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

data = list(zip(x, y))

kmeans = KMeans(n_clusters= 2)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```

```python
# clustered data with centeroids

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

data = list(zip(x, y))

kmeans = KMeans(n_clusters= 2)
kmeans.fit(data)

labels = kmeans.labels_
centeroids = kmeans.cluster_centers_

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=labels, cmap='viridis', marker = 'o', label='Data points')
plt.scatter(centeroids[:,0], centeroids[:,1], c='red', marker='X', s= 100, label='Centeroids')
plt.legend()
plt.title('Clusterd Data with Centeroids')
plt.show()
```

```python
# Interia lines for Each Point

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

data = list(zip(x, y))

kmeans = KMeans(n_clusters= 2)
kmeans.fit(data)

labels = kmeans.labels_
centeroids = kmeans.cluster_centers_

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c =labels, cmap='viridis', marker='o', label='Data Points')
plt.scatter(centeroids[:,0], centeroids[:,1], c='red', marker='X', s=200,label='Centeroids')

for i, point in enumerate(data):
  plt.plot([point[0], centeroids[labels[i], 0]], [point[1], centeroids[labels[i], 1]], 'k--', linewidth=0.5)

plt.legend()
plt.title('Clustered Data with Inertia Lines')
plt.show()
```
