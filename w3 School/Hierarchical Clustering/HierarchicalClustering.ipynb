```python
#Now we compute the ward linkage using euclidean distance, and visualize it using a dendrogram:

import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

x = [4, 5, 10, 4, 3, 11, 14, 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

data = list(zip(x, y))

linkage_data = linkage(data, method='ward', metric ='euclidean')
dendrogram(linkage_data)

plt.show()
```

```python
#Here, we do the same thing with Python's scikit-learn library. Then, visualize on a 2-dimensional plot:
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

# Sample data
x = [4, 5, 10, 4, 3, 11, 14, 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

data = list(zip(x, y))

# Perform Agglomerative Clustering
hierarchical_cluster = AgglomerativeClustering(n_clusters=2, linkage='ward')  # Remove affinity
labels = hierarchical_cluster.fit_predict(data)

# Plot the data points with colors based on cluster labels
plt.scatter(x, y, c=labels, cmap='viridis')  # Use colormap for better visual distinction
plt.title('Hierarchical Clustering')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(label='Cluster Label')  # Optional: add a color bar to show cluster labels
plt.show()
```

```python
#Pair Plot (Scatterplot Matrix)

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

x = [4, 5, 10, 4, 3, 11, 14, 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

labels = ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B']

# Sample data
data = pd.DataFrame({
    'X': x,
    'Y': y,
    'Label': labels
})

# Create a pair plot with a title
g = sns.pairplot(data, hue='Label', palette='viridis')
g.fig.suptitle('Pair Plot of Clusters', y=1.02)  # Adjust y for positioning
plt.show()

```

```python
#Dendrogram with Color Thresholds

from mpl_toolkits.mplot3d import Axes3D

# Sample data with a third dimension (for illustration)
z = [10, 11, 12, 10, 9, 15, 14, 12, 11, 10]  # Example third dimension
data = list(zip(x, y, z))

# Perform Agglomerative Clustering
hierarchical_cluster = AgglomerativeClustering(n_clusters=2, linkage='ward')
labels = hierarchical_cluster.fit_predict(data)

# Plot in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=labels, cmap='viridis')

ax.set_title('3D Hierarchical Clustering')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```
